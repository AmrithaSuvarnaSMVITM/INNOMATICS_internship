# Problem 2: Employee ID Card System


```python
class Employee_id_card:
    def __init__(self,emp_name,emp_id,emp_dept):
        self.emp_name= emp_name # employee name
        self.emp_id = emp_id # employee ID
        self.emp_dept = emp_dept # employee department

    # Method to display the employee ID card details
    def display_id_card(self):
        print("Employee ID_Card")
        print(f"Employee Name: {self.emp_name}")
        print(f"Employee ID: {self.emp_id}")
        print(f"Department: {self.emp_dept}")

emp_name = "Rahul"
emp_id = "EMP102"
emp_dept = "IT"

employee_id_card = Employee_id_card(emp_name,emp_id, emp_dept)
employee_id_card.display_id_card()
```

# Problem 3: Mobile Contact Record


```python
class Mobile_contact:
    def __init__(self,contact_name,phone_no):
        self.contact_name = contact_name # contact name
        self.phone_no= phone_no # contact phone number

    # Method to display the contact information
    def display_contact(self):
        print("Contact Saved") # Message confirmation- contact is saved
        print(f"Name: {self.contact_name}")
        print(f"Phone:{self.phone_no}")

contact_name="Anita"
phone_no="9876543210"

mobile_contact = Mobile_contact(contact_name,phone_no)
mobile_contact.display_contact()
```

# Problem 4: Product Price Tag Generator


```python
class Product_price_tag:
    def __init__(self,product_name,prd_price):
        self.product_name=product_name      # product name
        self.prd_price=prd_price    # product price

    # Method to display the formatted price tag
    def display_price(self):
        print(f"Product:{self.product_name}")
        print(f"Price: ₹{self.prd_price}")

product_name="Headphones"
prd_price=2499

product_price_tag = Product_price_tag(product_name, prd_price)
product_price_tag.display_price()
```

# Problem 5: Movie Rating Display System


```python
class Movie_rating_system:
    def __init__(self,movie_name,mov_rating):
        self.movie_name =movie_name # movie name
        self.mov_rating = mov_rating # movie rating

    # Method to display the movie details
    def display_movie_detail(self):
        print(f"Movie: {self.movie_name}")
        print(f"Rating: {self.mov_rating}/5") # rating out of 5

movie_name = "Inception"
mov_rating = 4.8

movie_rating_system = Movie_rating_system(movie_name, mov_rating)
movie_rating_system.display_movie_detail()
```
