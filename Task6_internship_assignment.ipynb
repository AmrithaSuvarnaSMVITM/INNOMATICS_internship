# Problem Statement 1:Smart Parking Lot Management System




```python
def smart_parking_sys(entry,exit,capacity):
  # total parked vehicles
    parked = entry - exit
    if parked < 0:
        parked = 0

    peak_usage = entry
    # parking status
    if parked > capacity:
        status = "Space Exceeded"
    else:
        status = "Available"
    return parked,peak_usage,status

parking_capacity = 50
entry = 5
exit = 2

parked,used,status = smart_parking_sys(entry,exit,parking_capacity)
print(f"Total Parked Vehicle:{parked}")
print(f"Peak Parking Usage:{used}")
print(f"Parking Status:{status}")
```

#  Problem Statement 2:Online Food Delivery Time Estimator


```python
def online_food_delivery(distance,traffic_level,condition):
    base_time = distance * 2.5

    # traffic delays
    if traffic_level.lower()=="low":
        t_delay = 0
    elif traffic_level.lower()=="medium":
        t_delay = 10
    else:  # High traffic
        t_delay = 20
    # weather delay
    if condition.lower() == "clear":
        w_delay = 0
    elif condition.lower() == "rainy":
        w_delay = 15
    else:  # Stormy or severe
        w_delay = 25

    final_eta = base_time + t_delay + w_delay
    return final_eta

distance = 8
traffic_level = "High"
condition = "Rainy"

delivery = online_food_delivery(distance,traffic_level,condition)

print(f"Estimated Delivery Time: {delivery} minutes")
```

#  Problem Statement 3:Movie Theatre Seat Occupancy Analyzer


```python
def seat_occupancy_analyzer(total_seats,occupied):
    # total booked seats
    count = len(occupied)

    # occupancy percentage
    occupancy_percentage = (count /total_seats)* 100

    if count == total_seats:
        status = "Housefull"
    elif occupancy_percentage >= 75:
        status = "Almost Full"
    else:
        status = "Available"

    if occupancy_percentage >= 90:
        suggest = "Additional shows available"
    else:
        suggest = "No additional shows. Sorry for inconvenience"
    return occupancy_percentage,status,suggest

total_seats = 200
occupied = [1]*150
occupance,status,suggest =seat_occupancy_analyzer(total_seats,occupied)
print(f"Occupancy: {occupance:.0f}%")
print(f"Show Status: {status}")
print(f"Suggestion: {suggest}")
```

#  Problem Statement 4:Cloud Server Load Classification System


```python
def cloud_server_load(cpu_usage):
    # average CPU usage
    average_cpu = sum(cpu_usage) / len(cpu_usage)

    if average_cpu < 50:
        status = "Normal"
    elif average_cpu <= 80:
        status = "Warning"
    else:
        status = "Critical"
    return average_cpu, status

cpu_readings = [45, 60, 70, 85, 90]
average_cpu, status = cloud_server_load(cpu_readings)

print(f"Average CPU Load:{average_cpu:.0f}%")
print(f"Server Status:{status}")
```

# Problem Statement 5:Smart Classroom Resource Usage Monitor


```python
def classroom_resource_usage(Resource_usage):
    overused_re = []

    usage = {
        "Projector": 5,
        "AC": 9,
        "Lights": 6
    }

    # Check each resource against Resource_usage
    for resource,n in Resource_usage.items():
        if resource in usage and n >usage[resource]:
            overused_re.append(resource)
    if overused_re:
        status = "Yes"
    else:
        status = "No"
    return overused_re,status

Resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

overused_resources,energy_alert = classroom_resource_usage(Resource_usage)
if overused_resources:
    print(f"Overused Resources:{overused_resources}")
else:
    print("Overused Resources: None")
print("Energy Alert:", energy_alert)
```

#  Problem Statement 6: Online Event Registration Capacity Controller

```python
def online_event_registration(capacity,track_registr):
    # registrations cannot exceed capacity
    if track_registr<= capacity:
        confirmed = track_registr
        list = 0
        status = "Open"
    else:
        confirmed = capacity
        list = track_registr - capacity
        status = "Closed"
    return confirmed,list,status

capacity = 100
track_registr = 105
confirmed,list,status= online_event_registration(capacity,track_registr)
print(f"Confirmed Registrations:{confirmed}")
print(f"Waitlisted Users:{list}")
print(f"Registration Status:{status}")
```
