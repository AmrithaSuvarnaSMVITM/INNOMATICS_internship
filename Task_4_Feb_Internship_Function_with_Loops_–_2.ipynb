# Problem 1: Mobile Recharge Validation System


```python
# given
def recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        try:
            amount= int(input("Enter recharge amount: ₹"))
            # minimum recharge amount
            if amount < 50:
                print("Invalid: Amount must be at least ₹50.Retry.\n")
                continue
            # Check if amount is one of the valid plans
            if amount in valid_plans:
                print(f"Recharge successful! Plan ₹{amount} activated.")
                break
            else:
                print("Invalid. Try again.\n")

        except ValueError:
            print("Invalid. Please re-enter a number.\n")
recharge()
```

# Problem 2: Inventory Reorder Alert System


```python
# given
def look_inventory(stock_dict):
    for product,stock in stock_dict.items():
        if stock < 15:
            print(f"{product}: Reorder Alert (Stock = {stock})")
        else:
            print(f"{product}: Stock OK (Stock = {stock})")
# dictionary of products and stock
inventory = {
    "Apples": 10,
    "Oranges": 25,
    "Bananas": 14,
    "Mangoes": 30
}

look_inventory(inventory)
```

# Problem 3: Student Result Processing System


```python
# given
def student_result(marks):
    total = 0
    for n in marks:
        total += n

    average =total /len(marks)

    if average >=50: # check pass/fail
        return f"Average Marks: {average:.2f} → Pass"
    else:
        return f"Average Marks: {average:.2f} → Fail"

student_marks = [60, 45, 70, 55]
result = student_result(student_marks)
print(result)
```

# Problem 4: Cab Fare Estimator with Retry Option


```python
# given
def cab_fare(distance, peak_hour=False):
    base_fare = 50
    per_km_rate = 12

    # Calculate distance charge
    distance_charge = distance * per_km_rate
    # Normal fare
    price = base_fare + distance_charge
    # Add 25% extra if peak hour (only on distance charge)
    if peak_hour:
        price = base_fare + (distance_charge * 1.25)
    return price

while True:
    try:
        distance = float(input("Enter distance travelled (in km): "))
        peak = input("Peak hour? (yes/no): ").strip().lower()

        peak_hour = True if peak == "yes" else False

        total_fare = cab_fare(distance, peak_hour)
        print(f"Estimated Fare: ₹{total_fare:.2f}")
        break

    except ValueError:
        print("Invalid input.\n")
```

# Problem 5: Employee Attendance Eligibility Checker


```python
# given
def elig_attendance(attendance_list):
    # Count present days
    present_days = 0
    for n in attendance_list:
        if n == "P":
            present_days += 1

    total_days = len(attendance_list)
    percentage =(present_days /total_days)*100

    if percentage >= 75:
        return f"Attendance: {percentage:.2f}% → Eligible"
    else:
        return f"Attendance: {percentage:.2f}% → Not Eligible"

attendance = ["P", "A", "P", "P", "A", "P", "P", "P"]
st_result = elig_attendance(attendance)
print(st_result)
```

# Problem 6: Password Strength Checker


```python
# given
def password_strength(password):
    # Minimum length
    if len(password) < 8:
        return "Weak: Password must have at least 8 characters.", False

    # Must contain at least one digit
    digits = False
    # Must contain at least one special character (@, #, $)
    special_char = False

    for ch in password:
        if ch.isdigit():
            digits = True
        if ch in "@#$":
            special_char = True

    if digits and special_char:
        return "Strong Password.", True
    elif not digits and not special_char:
        return "Weak: Must contain at least one digit and one special character (@, #, $).", False
    elif not digits:
        return "Weak: Must contain at least one digit.", False
    elif not special_char:
        return "Weak: Must contain at least one special character (@, #, $).", False


# Retry loop until strong password
while True:
    user_password = input("Enter your password: ")
    message, is_strong = password_strength(user_password)
    print(message)

    if is_strong:
        break   # exit loop once strong password is entered
```
