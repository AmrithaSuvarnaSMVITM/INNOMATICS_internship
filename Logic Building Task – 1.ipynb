# 1. User Login Check


```python
# given
username = "admin"
password = "1234"

# user input
ui = input("Enter the username: ")
pi = input("Enter the password: ")

# if-else loop
if ui == username and pi == password:
    print("Login Successful.")

else:
    print("Invalid.")
```

# 2. Pass / Fail Analyzer


```python
# given
marks = [45, 78, 90, 33, 60]

# initialize
pass_ct = 0
fail_ct = 0

# if-else loop
for n in marks:
    if n >=50:
        pass_ct +=1
    else:
        fail_ct +=1

print("Passed Students Count:", pass_ct)
print("Failed Students Count:", fail_ct)
```

# 3. Simple Data Cleaner

```python
# given
names = [" Alice ", "bob", " CHARLIE "]

cleaned_data = [name.strip().lower() for name in names]
# strip() removes leading and trailing spaces.
# lower() converts everything to lowercase.

print("Cleaned Names:", cleaned_data)
```

# 4. Message Length Analyzer

```python
# given
messages = ["Hi", "Welcome to the platform", "OK"]

for mssg in messages:
    length = len(mssg) # len is used to find the number of character in strings or items in list.
    print(f"Message: '{mssg}'")
    print(f"Length: {length}") # f is used to put variables directly inside the text

    # flag if longer than 10 characters
    if length > 10:
        print(f"'{mssg}'[Flagged]\n")
    else:
        print(f"'{mssg}'[OK]\n")
```

# 5. Error Message Detector


```python
# given
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_ct = 0

# if-else loop
for er in logs:
    if er == "ERROR":
        error_ct +=1  # Add 1 to the count

print(f"Total Error Count: {error_ct}")
```
