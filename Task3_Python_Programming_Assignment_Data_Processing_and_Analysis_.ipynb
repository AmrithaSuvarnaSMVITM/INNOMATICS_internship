# Problem Statement 1: Employee Performance Bonus Eligibility



```python
# given
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# find the max score
max_score = max(employees.values()) # gives value
# gives employees with max score
top_performance = []

for name, score in employees.items(): # gives both the key and value from the dictionary
    if score == max_score:
        top_performance.append(name) # add name at the end
# 'join' is used to join two strings
print("Top Performers Eligible for Bonus:",",".join(top_performance), f"(Score: {max_score})")
```

# Problem Statement 2: Search Query Keyword Analysis

```python
# given
query_key= "Buy mobile phone buy phone online"

# lowercase
query_key = query_key.lower()

import string
for p in string.punctuation: # module that contains all common punctuation marks
    query_key = query_key.replace(p, "") # Replaces part of a string with something else

# split into words
token = query_key.split()
# word frequency count
freq = {}
for word in token:
    if word in freq:
        freq[word]+=1
    else:
        freq[word]=1

result = {}
for word, count in freq.items():
    if count > 1:          # check condition
        result[word] = count   # add to result dictionary

print(result)
```

# Problem Statement 3: Sensor Data Validation


```python
# given
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

reading = []

for hour, data in enumerate(sensor_readings):
    if data % 2 == 0: # check if even
        reading.append((hour, data))

print("Valid Sensor Readings (Hour, Value):")
print(reading)
```

# Problem Statement 4: Email Domain Usage Analysis


```python
# given
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

email_domains = []
for email in emails:
    email_domain = email.split("@")[1]
    email_domains.append(email_domain)

# Count frequency
domain_count = {}
for n in email_domains:
    if n in domain_count:
        domain_count[n]+=1
    else:
        domain_count[n]=1

# percentage usage
total_users = len(emails)
for d, count in domain_count.items():
    percentage = (count / total_users) * 100
    print(f"{d}:{int(percentage)}%")
```

# Problem Statement 5: Sales Spike Detection

```python
# given
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average daily sales
average_sales = sum(sales) / len(sales)

incr = []
for n, data in enumerate(sales, start=1):
    if data > average_sales * 1.3 or data == 2200:
        incr.append((n, data))

for n, data in incr:
    print(f"Day {n}: {data}")
```

# Problem Statement 6: Duplicate User ID Detection


```python
# given
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
# Count of each user ID
freq = {}
for u_id in user_ids:
    if u_id in freq:
        freq[u_id]+=1
    else:
        freq[u_id]=1

# duplicates
duplicates = {}
for u_id, n in freq.items():
    if n > 1:
        duplicates[u_id] = n

for u_id, n in duplicates.items():
    print(f"{u_id} → {n} times")
```
