# Problem 1: Unique Words in a Sentence

```python
# given
message = "python is easy and python is powerful"

# spliting words
words = message.split()

# remove duplicates
uq_words = set(words)

print("Unique words count:", len(uq_words))
print("Unique words:", uq_words)
```

# Problem 2: Highest Salary from Employee Data


```python
# given
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

high_empl= max(employees, key=employees.get) # 'get' is used to get the value of a key
high_sal = employees[high_empl]

print("Highest Salary:", high_empl, ":", high_sal)
```

# Problem 3: Find Maximum and Minimum Values


```python
# given
numbers = [45, 22, 89, 10, 66]

# initialize
max_value = numbers[0]
min_value = numbers[0]

for n in numbers:
    if n > max_value:
        max_value = n
    if n < min_value:
        min_value = n

print("List:", numbers)
print("Max:", max_value)
print("Min:", min_value)
```

# Problem 4: Count Products Above a Price Threshold


```python
# given
prices = [450, 1200, 899, 1500, 300]
count = 0

for n in prices:
    if n > 1000:
        count += 1

print("Products above 1000:", count)
```

# Problem 5: Calculate Attendance Percentage


```python
# given
attendance = ["P", "P", "A", "P", "P"]

# 'count' is used to count the number of letters
present_ct = attendance.count("P")

total = len(attendance)
attendance_perct = (present_ct / total) * 100

print("Attendance Percentage:", attendance_perct)
```

# Problem 7: Count Character Frequency


```python
text = "alphabet"

# dictionary
char_frq = {}

#loop
for n in text:
    if n in char_frq:
        char_frq[n] += 1
    else:
        char_frq[n] = 1

print(char_frq)
```

# Problem 9: Check if a Key Exists in a Dictionary


```python
# given
department = {
    "Employee": "Sam",
    "Salary": 85000,
    "Department": "IT"
}
# loops
if "Employee" in department:
    print("Employee exists")
else:
    print("Employee does not exist")
```

# Problem 10: Calculate Average Marks


```python
marks = [80, 70, 65, 75]

total = sum(marks) # sum
count = len(marks) # length

avg = total/count # average

print("Average Marks:",avg)
```
