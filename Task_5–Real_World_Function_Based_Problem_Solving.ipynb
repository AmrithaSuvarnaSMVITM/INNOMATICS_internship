# Problem 1: Social Media – Post Engagement Analyzer

```python
# given
def post_engagement(likes_list):
    total_likes =0
    for like in likes_list:
        total_likes +=like

    if total_likes >= 1000:
        analyze ="Viral Post"
    else:
        analyze ="Normal Engagement"

    return total_likes, analyze

data = [200, 300, 150, 400]   # likes from different sources
total,analyze =post_engagement(data)

print(f"Total Likes: {total}")
print(f"Post Status: {analyze}")
```

# Problem 3: Agriculture – Rainfall Adequacy Checker


```python
# given
def rainfall_adequacy_checker(data, required_level):
    total = 0
    for n in data:
        total +=n
    average = total /len(data)

    # Check adequacy
    if average >= required_level:
        weather = "Adequate Rainfall"
    else:
        weather = "Inadequate Rainfall"
    return average,weather

rainfall = [80, 60, 70, 78]
required_level = 70                    # required average rainfall level

average, status = rainfall_adequacy_checker(rainfall, required_level)

print(f"Average Rainfall:{average:.0f}")
print(f"Rainfall Status: {status}")
```

# Problem 4: Social Media – Duplicate Account Detection


```python
# given
def duplicate_detect(usernames):
    unique = set()
    duplicates = set()

    for user in usernames:
        if user in unique:
            duplicates.add(user)   # flagging duplicate
        else:
            unique.add(user)

    # Check if duplicates exist
    if duplicates:
        print("Flagged Accounts:", ", ".join(duplicates))
    else:
        print("Accounts are unique.")

usernames_list = ["alice", "bob", "charlie", "alice", "david", "bob"]
duplicate_detect(usernames_list)
```

# Problem 6: Agriculture – Premium Crop Price Filter

```python
# given
def filter_premium_crop(crop_prices):
    premium_crops =[]

    for price in crop_prices:
        if price > 2000:
            premium_crops.append(price)
    return premium_crops

crop_prices =[1500, 2500, 1800, 3200, 1100]
filter =filter_premium_crop(crop_prices)

print(f"Premium Crops:{filter}")
```

# Problem 7: System Monitoring – Application Health

```python
# given
def check_monitor_health(Errors):
    if Errors == 0:
        status = "Healthy"
    elif Errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    return Errors,status

error_list = [7,0,5,3]
for e in error_list:
    count, status = check_monitor_health(e)
    print(f"Error Count:{count}")
    print(f"System Status: {status}")
```

# Problem 9: E-Learning – Student Attendance Eligibility System


```python
# given
def check_attendance_elig(att_records, total_classes):
    for Attendance in att_records:
        # Calculate attendance percentage
        percentage = (Attendance /total_classes) * 100

        # Check eligibility
        if Attendance >=75:
            status ="Eligible"
        else:
            status ="Not Eligible"

        print(f"Attendance Percentage:{percentage:.1f}")
        print(f"Exam Eligibility: {status}\n")

data = [40, 30, 25]   # attended class for different students
total_classes = 50    # total classes

check_attendance_elig(data, total_classes)
```
